## Criando mapas de calor anotados 
Será necessário um matriz 2D ou matriz que contenha os dados para o código de cores do mapa de calor.
Também será necessário duas listas ou matrizes de categorias para os eixos x e y.  

É importante definir os locais dos ticks ( set_xticks) e também os rótulos dos ticks ( set_xticklabels), caso contrário, eles ficariam fora de sincronia. 

Os locais são apenas os números inteiros crescentes, enquanto os ticklabels são os rótulos a serem mostrados. 

Finalmente, podemos rotular os próprios dados criando um Textdentro de cada célula mostrando o valor dessa célula.




```python
matrizmesregiao
```

```python
matrizmesregiao[1:, 1:]
```

```python
plt.rcParams["figure.figsize"] = [6, 12]
regioes = matrizmesregiao[0 , 1:12]
meses = matrizmesregiao[1:12, 0]

dadoscovidregiao =  matrizmesregiao[1:12, 1:].astype('i')


fig, ax = plt.subplots()
im = ax.imshow(dadoscovidregiao, cmap = 'magma_r')

# We want to show all ticks...
ax.set_xticks(np.arange(len(regioes)))
ax.set_yticks(np.arange(len(meses)))


# # ... and label them with the respective list entries
ax.set_xticklabels(regioes)
ax.set_yticklabels(meses)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(meses)):
    for j in range(len(regioes)):
        text = ax.text(j, i, dadoscovidregiao[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Dados da Covid por Região e Mês")

fig.tight_layout()

plt.show()
```

```python
plt.rcParams["figure.figsize"] = [10, 15]

estados = matrizmesestado[0 , 1:]
meses = matrizmesestado[1:12, 0]

dadoscovidestados =  matrizmesestado[1:12, 1:].astype('i').T


fig, ax = plt.subplots()
im = ax.imshow(dadoscovidestados, cmap = 'magma_r')

# We want to show all ticks...
ax.set_yticks(np.arange(len(estados)))
ax.set_xticks(np.arange(len(meses)))


# ... and label them with the respective list entries
ax.set_yticklabels(estados)
ax.set_xticklabels(meses)


# Turn spines off and create white grid.
for edge, spine in ax.spines.items():
    spine.set_visible(False)

ax.set_xticks(np.arange(dadoscovidestados.shape[1]+1)-.5, minor=True)
ax.set_yticks(np.arange(dadoscovidestados.shape[0]+1)-.5, minor=True)
ax.grid(which="minor", color="k", linestyle='-', linewidth=3)
ax.tick_params(which="minor", bottom=False, left=False)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(estados)):
    for j in range(len(meses)):
        text = ax.text(j, i, dadoscovidestados[i, j],
                       ha="center", va="center", color="b")

ax.set_title("Dados da Covid por Estado e Mês")

fig.tight_layout()

plt.show()
```
