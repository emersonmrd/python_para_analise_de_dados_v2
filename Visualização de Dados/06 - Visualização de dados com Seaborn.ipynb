# Visualização de dados com Seaborn

### Carga de Arrays com dados das Covid para uso nos gráficos

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
```

```python
diratual = os.getcwd()
diretorio = os.path.join(diratual) 
os.listdir(diretorio)
```

```python
arquivo = 'total_brasil.csv'
caminho = os.path.join(diretorio,arquivo)
total_brasil = np.loadtxt(caminho, delimiter=';', skiprows=0 , dtype='U', encoding = 'utf-8')
print(total_brasil[:5])
```

# Gráficos com Seabor

```python
sns.set(style="white" )
plt.rcParams["figure.figsize"] = [6, 4]

x= total_brasil[:12,0]
y= total_brasil[:12,1].astype('i')/1000

fig = sns.lineplot(x=x, y=y, marker='o', linestyle='dashed')

plt.title('Evolução dos Casos de COVID-19 no Brasil (Em milhares)')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.ylim(bottom=0)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_theme(style="white")

xy = sorted(zip(totaisregioes_arr[:12,0], totaisregioes_arr[:12,1].astype('i')/1000), key=lambda pair: pair[1], reverse=True)
x= [x for x , _ in xy]
y= [y for _ , y in xy]
print(xy)

fig = sns.barplot(x=y, y=x, palette="rocket", hue=x, dodge=False, legend=False)
plt.title('Evolução dos Casos de COVID-19 no Brasil (Em milhares)')
plt.xlabel('Mês')
plt.ylabel('Quantidade')

plt.show()
```

```python
sns.set_theme(style="white" )

xy = sorted(zip(totaisregioes_arr[:12,0], totaisregioes_arr[:12,1].astype('i')/1000), key=lambda pair: pair[1], reverse=True)
x= [x for x , _ in xy]
y= [y for _ , y in xy]
print(xy)
fig , (ax1, ax2) = plt.subplots(2, 1)

sns.barplot(x=x, y=y,palette="rocket", ax=ax1, hue=x)
ax1.set_title('Evolução dos Casos de COVID-19 no Brasil (Em milhares)')
#ax1.set_xlabel('Região')
ax1.set_ylabel('Quantidade')
ax1.set_ylim(bottom=0)

sns.barplot(x=y, y=x,palette="rocket", ax=ax2, hue=x , orient='horizontal', dodge=False)
ax2.set_xlabel('Quantidade')
ax2.set_ylabel('Região')

plt.show()
```

```python
plt.rc('figure', figsize = (8.27, 11.7))
xy = sorted(zip(totaisregioes_arr[:12,0], totaisregioes_arr[:12,1].astype('i')/1000), key=lambda pair: pair[1], reverse=True)
labels= [x for x , _ in xy]
sizes= [y for _ , y in xy]
print(xy)

fig , (ax1, ax2) = plt.subplots(2, 1)

theme = plt.get_cmap('copper')
cores = [theme(1 * i / len(sizes)) for i in range(1,len(sizes)+1)]
ax1.set_title('Proporção de Casos de COVID-19 no Brasil por Região')
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', 
        colors =cores,  textprops={'color':"w"})
ax1.axis('equal')
ax1.legend(labels,
          loc="lower left")


ax2.pie(sizes, labels=labels, autopct='%1.1f%%', 
        colors =cores,  textprops={'color':"w"},
        wedgeprops=dict(width=0.6))
ax2.axis('equal')


plt.show()
```

```python
plt.rc('figure', figsize = (14, 10))

x= np.log(totaisestados_arr[:,1].astype(np.int64))
y= np.log(totaisestados_arr[:,2].astype(np.int64))
l = totaisestados_arr[:,0]

print(xy)
fig , (ax1, ax2) = plt.subplots(2, 1)

sns.scatterplot(x=x, y=y, ax=ax1)
ax1.set_title('Casos X Obitos por COVID-19 no Brasil (Em milhares)')
#ax1.set_xlabel('Região')
ax1.set_ylabel('Log(Quant. Óbitos)')
#ax1.set_xlabel('Log(Quant. Casos)')

for l1, x1, y1 in zip(l, x, y):
  ax1.annotate(l1, xy=(x1,y1),
               xytext=(5,-5),
               textcoords='offset points')
  
sns.regplot(x=x, y=y, ax=ax2)
ax2.set_ylabel('Log(Quant. Óbitos)')
ax2.set_xlabel('Log(Quant. Casos)')

for l1, x1, y1 in zip(l, x, y):
  ax2.annotate(l1, xy=(x1,y1),
               xytext=(5,-5),
               textcoords='offset points')
plt.show()
```
